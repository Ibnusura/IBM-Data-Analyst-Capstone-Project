## Demo: How to work with database


Download database file.


```python
import requests
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite"

response = requests.get(url)
with open("m4_survey_data.sqlite", "wb") as file:
    file.write(response.content)
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
QUERY ="""
select ConvertedComp
from master
"""

df = pd.read_sql_query(QUERY, conn)
sns.displot(df, height=7, aspect=1.5)
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY="""
select Age
from master
"""

df = pd.read_sql_query(QUERY,conn)
print(df)

plt.rcParams['figure.figsize'] = [15,8]
ax = sns.boxplot(df, orient='h')
ax.set_title('Age Distribution', size=18)

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY ="""
select Age, WorkWeekHrs
from master
"""

df = pd.read_sql_query(QUERY,conn)
print(df)

sns.scatterplot(df, x='Age', y='WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
select WorkWeekHrs, CodeRevHrs, Age
from master
"""

df = pd.read_sql_query(QUERY,conn)
print(df)

sns.scatterplot(df, x='CodeRevHrs', y='WorkWeekHrs', size='Age', alpha=0.5, sizes=(1,1000))
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
select *, count(*) as count
from DatabaseDesireNextYear
group by DataBaseDesireNextYear
order by count desc
limit 5
"""

df = pd.read_sql_query(QUERY, conn)
print(df)

data = df['count']
labels = df['DatabaseDesireNextYear']

fig, ax = plt.subplots()
ax.pie(data, labels=labels, autopct='%1.1f%%', explode=(0.1,0,0,0,0), shadow=True)
plt.show()
```

Popular language respondents wish to learn next year

```python
QUERY = """
select *, count(*) as count
from LanguageDesireNextYear
group by LanguageDesireNextYear
order by count desc
limit 5
"""

df = pd.read_sql_query(QUERY, conn)
print(df)
```

Respondents that currently work with 'SQL'

```python
QUERY = """
select *, count(*) as count
from LanguageWorkedWith
group by LanguageWorkedWith
order by count desc
limit 5
"""

df = pd.read_sql_query(QUERY, conn)
print(df)
```

```python
# Query to get the data from the database
QUERY = """
SELECT Respondent, DatabaseWorkedWith 
FROM DatabaseWorkedWith
ORDER BY Respondent
"""
df = pd.read_sql_query(QUERY, conn)

# Group the data by 'Respondent' and aggregate the unique databases they worked with
respondent_db_count = df.groupby('Respondent')['DatabaseWorkedWith'].nunique()

# Filter respondents who worked with only 1 database
single_db_respondents = respondent_db_count[respondent_db_count == 1].index.tolist()  # Convert index to list

# Filter the dataframe for respondents who worked with only 'MySQL'
uniq_mysql = df[(df['Respondent'].isin(single_db_respondents)) & (df['DatabaseWorkedWith'] == 'MySQL')]

# Count the number of respondents who worked with 'MySQL' only
mysql_only_count = uniq_mysql['Respondent'].nunique()

print(mysql_only_count)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
select WorkWeekHrs, CodeRevHrs, Age
from master
where Age between 30 and 50
"""

df = pd.read_sql_query(QUERY, conn)
print(df)

work_week = df['WorkWeekHrs'].quantile(0.5)
code_rev = df['CodeRevHrs'].quantile(0.5)

plt.rcParams['figure.figsize'] = [6,6]

fig, ax = plt.subplots()
ax.bar('Age 30 - 50', work_week, label='WorkWeekHrs')
ax.bar('Age 30 - 50', code_rev, bottom=work_week, label='CodeRevHrs')
ax.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
select ConvertedComp, Age
from master
where Age between 45 and 60
"""

df = pd.read_sql_query(QUERY, conn)

median = df.groupby('Age')['ConvertedComp'].median()

data = pd.DataFrame(list(zip(median.index, median.values)), columns=['Age', 'ConvertedComp'])

sns.lineplot(data, x='Age', y='ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
select MainBranch, count(*) as freq
from master
group by MainBranch
"""

df = pd.read_sql_query(QUERY, conn)
sns.barplot(df, x='freq', y='MainBranch', orient='h')

plt.show()
```

```python
conn.close()
```
